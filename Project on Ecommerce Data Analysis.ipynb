```python
def databaseConnect():
    mydb=mysql.connector.connect(
        host="localhost",
        user="root",
        password="root",
        database="ecommerce_db"
    )
    return mydb   
```

### 2.Query Data From The 'customer', 'product' & 'order_details' Tables

```python
conn=databaseConnect()
sql="SELECT * FROM customer"
customerFrame=pd.read_sql(sql,conn)
print(customerFrame)

sql="SELECT * FROM product"
productFrame=pd.read_sql(sql,conn)
print(productFrame)

sql="SELECT * FROM order_details"
orderFrame=pd.read_sql(sql,conn)
print(orderFrame)
```

### 3.Removing Missing Values & Duplicate Values

```python
conn=databaseConnect()
sql="SELECT * FROM customer"
customerFrame=pd.read_sql(sql,conn)
customerFrame.dropna(inplace=True)
customerFrame.drop_duplicates(inplace=True)

sql="SELECT * FROM product"
productFrame=pd.read_sql(sql,conn)
productFrame.dropna(inplace=True)
productFrame.drop_duplicates(inplace=True)

sql="SELECT * FROM order_details"
orderFrame=pd.read_sql(sql,conn)
orderFrame.dropna(inplace=True)
orderFrame.drop_duplicates(inplace=True)

print("After fixing data.....")
print(customerFrame)
print(productFrame)
print(orderFrame)
```

## Exploratory Data Analysis (EDA) and Visualization:

### A. Customer Analysis:

#### 1.Identify the Total Number of Customers City wise.


```python
mycolor=np.array(['red','blue','yellow','green'])
customerFrame['name'].groupby(customerFrame['city']).size().plot(kind='bar', color=mycolor)
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.title(' City-wise Number of Customers')
plt.show()
```

### B. Product Analysis:

#### 1.Identify the Total Number of Products Category wise.

```python
mycolor=np.array(['cyan','yellow','pink'])
productFrame['product_name'].groupby(productFrame['category']).size().plot(kind='bar',color=mycolor)
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.title('Category-wise Products')
plt.show()
```

#### 2.Identify the Total Number of Products Sub-category wise.

```python
mycolor=np.array(['red','purple','blue','orange','yellow','magenta','green'])
productFrame['product_name'].groupby(productFrame['sub_category']).size().plot(kind='bar',color=mycolor)
plt.xlabel('Sub Category')
plt.ylabel('Number of Products')
plt.title('Sub Category-wise Products')
plt.show()
```

#### 3.Identify the Percentage of Products Stock wise.

```python
productFrame.groupby('product_name')['stock'].sum().sort_values(ascending=False).plot(kind='pie', autopct='%1.1f%%')
plt.title('Stock-wise Products')
plt.show()
```

#### 4.Identify the Percentage of Products Selling Price wise.

```python
productFrame.groupby('product_name')['selling_price'].sum().sort_values(ascending=False).plot(kind='pie', autopct='%1.1f%%')
plt.title('Selling Price-wise Products')
plt.show()
```

#### 5.Identify the Stock-level of Products Category wise.

```python
mycolor=np.array(['red','blue','yellow',])
productFrame['stock'].groupby(productFrame['category']).size().plot(kind='bar',color=mycolor)
plt.xlabel('Category')
plt.ylabel('Stock Level')
plt.title('Category-wise Stock Level of Products')
plt.show()
```

#### 6.Identify the Original Price vs Selling Price for Each Product.

```python
plt.plot(productFrame['product_name'], productFrame['original_price'], marker='o', linestyle='-', label='Original Price')
plt.plot(productFrame['product_name'], productFrame['selling_price'], marker='o', linestyle='-', label='Selling Price')
plt.xlabel('Product Name')
plt.ylabel('Price')
plt.title('Original Price vs Selling Price for Each Product')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

#### 7.Identify the Difference Between Original Price and Selling Price for Each Product.

```python
productFrame['price_difference'] = productFrame['original_price'] - productFrame['selling_price']
positive_difference_percentage = (productFrame['price_difference'] > 0).mean() * 100
negative_difference_percentage = (productFrame['price_difference'] < 0).mean() * 100

bars=plt.barh(productFrame['product_name'], productFrame['price_difference'], color='coral')
for bar, difference in zip(bars, productFrame['price_difference']):
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{difference:.2f}', va='center', ha='left')
plt.ylabel('Product Name')
plt.xlabel('Price Difference (Original Price - Selling Price)')
plt.title('Difference Between Original Price and Selling Price for Each Product')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 8.Calculate the Average, Maximum, and Minimum Selling Prices for Pproduct (In Scatterplot).
.

```python
average_price=productFrame['selling_price'].mean()
maximum_price=productFrame['selling_price'].max()
minimum_price=productFrame['selling_price'].min()
print(f"Average Price: {average_price}")
print(f" Maximum Price: {maximum_price}")
print(f" Minimum Price: {minimum_price}")

plt.figure(figsize=(10, 6))
plt.plot(productFrame['product_name'],productFrame['selling_price'], color='blue', marker='o', label='Selling Price')
plt.axhline(y=average_price,color='red', marker='o',label='Average Selling Price')
plt.axhline(y=maximum_price,color='orange', marker='o', label='Max Selling Price')
plt.axhline(y=minimum_price,color='green', marker='o', label='Min Selling Price')
plt.title('Average, Maximum & Minimum Selling Price of Products')
plt.xticks(rotation=90)
plt.xlabel('Product Name')
plt.ylabel('Selling Price')
plt.legend()
plt.grid(True)
plt.show()
```

#### 9.Calculate the Average, Maximum, and Minimum Selling Prices for Products (In Bar Diagram).

```python
plt.bar(['Average', 'Maximum', 'Minimum'], [average_price, maximum_price, minimum_price], color=['blue', 'green', 'orange'])
plt.title('Average, Maximum, and Minimum Selling Prices for Products')
plt.ylabel('Selling Price')
plt.show()
```

### C. Order Analysis:

#### 1.Identify the Total Number of Orders Product wise.

```python
df = pd.merge(productFrame, orderFrame, on='product_id')
df['order_id'].groupby(df['product_name']).size().plot(kind='bar',color='yellow')
plt.xlabel('Product Name')
plt.ylabel('Number of orders')
plt.title('Product-wise Total Number of Orders')
plt.show()
```

#### 2.Identify the Total Number of Orders Customer wise.

```python
df = pd.merge(customerFrame, orderFrame, on='customer_id')
df['order_id'].groupby(df['name']).size().plot(kind='bar',color='purple')
plt.xlabel('Customer Name')
plt.ylabel('Number of orders')
plt.title('Customer-wise Total Number of Orders')
plt.show()
```

#### 3.Identify the Total Number of Orders Payment Mode wise.

```python
orderFrame['order_id'].groupby(orderFrame['payment_mode']).size().plot(kind='bar',color=['cyan','pink'])
plt.xlabel('Payment Mode')
plt.ylabel('Number of orders')
plt.title('Payment Mode-wise Total Number of Orders')
plt.show()
```

#### 4.Identify the Total Number of Orders Order Status wise.

```python
orderFrame['order_id'].groupby(orderFrame['order_status']).size().plot(kind='bar',color=['magenta','orange'])
plt.xticks(rotation=90)
plt.xlabel('Order Status')
plt.ylabel('Number of orders')
plt.title('Order Status Distribution')
plt.show()
```

#### 5.Identify the Percentage of Orders Date wise.

```python
orderFrame.groupby('order_date')['order_id'].sum().sort_values(ascending=False).plot(kind='pie', autopct='%1.1f%%')
plt.title('Date wise Total Number of Orders')
plt.show()
```

#### 6.Identify the Percentage of Orders City wise.

```python
df = pd.merge(customerFrame, orderFrame, on='customer_id')
city_wise_orders = df.groupby('city')['order_id'].count().reset_index()
plt.pie(city_wise_orders['order_id'], labels=city_wise_orders['city'], autopct='%1.1f%%', startangle=140)
plt.title('City-wise Distribution of Orders')
plt.show()
```

### D. Sales Analysis:

#### 1.Calculate Total Revenue Generated from Orders Product wise.

```python
merged_df = pd.merge(productFrame, orderFrame, on='product_id')
merged_df['total_revenue'] = merged_df['quantity'] * merged_df['selling_price']
product_wise_revenue = merged_df.groupby('product_name')['total_revenue'].sum().reset_index()
plt.bar(product_wise_revenue['product_name'], product_wise_revenue['total_revenue'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.title('Total Revenue Generated from Orders (Product-wise)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 2.Calculate Total Revenue Product Category wise Percentage.

```python
merged_df = pd.merge(productFrame, orderFrame, on='product_id')
merged_df['total_revenue'] = merged_df['quantity'] * merged_df['selling_price']
category_wise_revenue = merged_df.groupby('category')['total_revenue'].sum().reset_index()
category_wise_revenue['percentage_of_total'] = (category_wise_revenue['total_revenue'] / category_wise_revenue['total_revenue'].sum()) * 100
plt.pie(category_wise_revenue['percentage_of_total'], labels=category_wise_revenue['category'], autopct='%1.1f%%', startangle=140)
plt.title('Total Revenue Percentage by Product Category')
plt.show()
```

#### 3.Identify the most profitable products based on the difference between
original and selling prices


```python
productFrame['profit'] = productFrame['original_price'] - productFrame['selling_price']
most_profitable_products = productFrame.sort_values(by='profit', ascending=False)
plt.bar(most_profitable_products['product_name'], most_profitable_products['profit'], color='purple')
plt.xlabel('Product Name')
plt.ylabel('Profit (Original Price - Selling Price)')
plt.title('Most Profitable Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 4.Identify Product Names with the Highest & Lowest Order Quantities.

```python
merged_df = pd.merge(productFrame, orderFrame, on='product_id')
product_order_quantities = merged_df.groupby('product_name')['quantity'].sum().reset_index()
sorted_products = product_order_quantities.sort_values(by='quantity', ascending=False)
highest_quantity_product = sorted_products.iloc[0]
lowest_quantity_product = sorted_products.iloc[-1]
plt.plot(sorted_products['product_name'], sorted_products['quantity'], marker='o', linestyle='-', color='green')
plt.xlabel('Product Name')
plt.ylabel('Total Order Quantity')
plt.title('Order Quantities for Each Product')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 5.Identify Customer Names with the Highest & Lowest Order Quantities.

```python
merged_df = pd.merge(customerFrame, orderFrame, on='customer_id')
customer_order_quantities = merged_df.groupby('name')['quantity'].sum().reset_index()
sorted_customers = customer_order_quantities.sort_values(by='quantity', ascending=False)
highest_quantity_customer = sorted_customers.iloc[0]
lowest_quantity_customer = sorted_customers.iloc[-1]
plt.plot(sorted_customers['name'], sorted_customers['quantity'], marker='o', linestyle='-', color='orange')
plt.xlabel('Customer Name')
plt.ylabel('Total Order Quantity')
plt.title('Order Quantities for Each Customer')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### 6.Calculate the Percentage of Products Sold & Not Sold.

```python
merged_df = pd.merge(productFrame, orderFrame, on='product_id', how='left')
merged_df['sold'] = merged_df['quantity'].notnull()
products_sold_percentage = merged_df['sold'].value_counts(normalize=True) * 100
plt.pie(products_sold_percentage, labels=products_sold_percentage.index.map({True: 'Sold', False: 'Not Sold'}),
        autopct='%1.1f%%', startangle=140, colors=['blue', 'coral'])
plt.title('Percentage of Products Sold and Not Sold')
plt.show()
```
